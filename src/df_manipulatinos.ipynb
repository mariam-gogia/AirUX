```python
# data structure and manipulation
import pandas as pd
import numpy as np 

# data visualization
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation

# files and time 
import csv
import sys
import time
import datetime as dt
```

```python
# import data
file = 'data.csv'
df = pd.read_csv(file)
df.head(100)
```

```python
time_steps = df['id']

angles_values = {
    "radius": df['r'], 
    "theta": df['theta'], 
    "phi": df['phi'],  
}

acceleration_values = {
    "acc_x": df['acc_x'], 
    "acc_y": df['acc_y'], 
    "acc_z": df['acc_z'],    
}

change_values = {
    "delta_r": df['delta_r'], 
    "delta_theta": df['delta_theta'], 
    "delta_phi": df['delta_phi'], 
    "delta_acc_x": df['delta_acc_x'], 
    "delta_acc_y": df['delta_acc_y'], 
    "delta_acc_z": df['delta_acc_z'], 
}
```

```python
# visualization functions
def visualize_separetly(values):
    for value in values:
        plt.plot(time_steps, values[value], linewidth=0.5, label=value)    
        plt.title("Signal Visualization", color="white")
        plt.xlabel("Time Steps", color='white')
        plt.ylabel(value, color="white")
        plt.legend()
        plt.show()
        
def visualize_together(values):
    for value in values:
        plt.plot(time_steps, values[value], linewidth=0.5, label=value)    
        plt.title("Signal Visualization", color="white")
        plt.xlabel("Time Steps", color='white')
        plt.ylabel(value, color="white")
        plt.legend()
    plt.show()
```

```python
# visualization
visualize_together(angles_values)
visualize_together(acceleration_values)
visualize_together(change_values)
```

```python
visualize_separetly(change_values)
```
